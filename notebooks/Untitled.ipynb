```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras import backend as K
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
x_train = []
y_train = []
lens = []
Len = 80
for i in ['flowerbed2.npy', 'flowerbed3.npy', 'flowerbed7.npy', 'flowerbed8.npy']:
    data = np.load(i,allow_pickle = True)
    print(len(data))
    #plt.plot(data[0][0])
    #print(data[0][2])

    for j in range(len(data)):
        if(len(data[j][0]) >Len):
            
            #while(len(data[j][0])<Len):
            #    data[j][0] = np.append(data[j][0], data[j][0][-1])
                
            
            for k in range(len(data[j][0])-Len):
                if(k<5):
                    y_train.append(np.array(data[j][0][1 + k:Len + k]))
                #(series, trenutna, temp, Äas?)
                
                    feature_vector = []
                    for l in range(20):
                        feature_vector.append(data[j][0][1+k+l])
                    feature_vector.append(data[j][2]) #temp
                    feature_vector.append(datetime.fromtimestamp(data[j][1][0]).hour) #hour
                    feature_vector.append(datetime.fromtimestamp(data[j][1][0]).day) #day
                
                    x_train.append(feature_vector)
                    if(np.isnan(data[j][0][0])):
                        y_train = y_train[:-1]
                        x_train = x_train[:-1]
            
            #data[j][2] -- temperatura!
            #x_train.append([data[j][0][0], data[j][2], datetime.fromtimestamp(data[j][1][0]).hour])
                    lens.append(len(y_train[-1]))
        

x_train = np.array(x_train)
y_train = np.array(y_train)
#print(x_train.shape)
#print(y_train.shape)

d = np.array([0, 0], dtype = object)
d[0] = x_train
d[1] = y_train

np.save('train_data_carouge.npy', d)

x_train /= 30
y_train /= 30          
#for i in range(len(y_train)):
#    plt.plot(y_train[i])
#print(lens) 
#print(str(len(x_train)) + ' samples')

shuffler = np.random.permutation(len(x_train))
x_train = x_train[shuffler]
y_train = y_train[shuffler]


print(x_train.dtype)
```
